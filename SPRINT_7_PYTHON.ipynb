## 2. Convertidor de temperatures

Existen diversas unidades de temperatura utilizadas en distintos contextos y regiones. Las más comunes son Celsius (°C), Fahrenheit (°F) y Kelvin (K). También existen otras unidades como Rankine (°Ra) y Réaumur (°Re).
 
 Selecciona al menos 2 conversores, de modo que al introducir una temperatura devuelva, como mínimo, dos conversiones, de modo que se puedan guardar (recuerda que un print() no se puede guardar nunca).
 
 Consejo: Intenta validar los datos previamente, para que envíe un mensaje de advertencia si los datos introducidos por el usuario/a no están en el formato adecuado.
 
 (EXTRA): Piensa una manera de almacenar todas las posibles conversiones en un solo objeto (¿Lista? ¿Diccionario? ¿DataFrame?) en lugar de escribir muchos if else en función de la temperatura de origen y la temperatura de destino.

```python
def conversor_temperaturas(centigrados):
    
    ''' Funcion que convierte los grados Celsius ingresados a Farenheit y Kelvin
        y los guarda en un historial de grados ya convertidos'''
    
    if not isinstance(centigrados, (int, float)):
        raise ValueError('La temperatura debe estar en número entero o decimal')

    farenheit = (centigrados * 9/5) + 32
    kelvin = centigrados + 273.15

    #diccionario de conversion
    resultado = {                          
        "Celsius": centigrados,
        "Fahrenheit": round(farenheit, 2),
        "Kelvin": round(kelvin, 2)
    }
    
    #Crea un atributo 'historial' con una lista vacia
    if not hasattr(conversor_temperaturas, "historial"):  
        conversor_temperaturas.historial = []

    # Suma cada resultado convertido al historial
    conversor_temperaturas.historial.append(resultado)
    
    # Cada vez que se ejecuta la funcion agrega el nuevo resultado al historial
    return conversor_temperaturas.historial
```

## 3. Comptador de paraules d'un text.
 
Escribe una función que dado un texto, muestre las veces que aparece cada palabra. 
Intenta que se gestionen todas las posibles casuísticas que hagan que el programa no funcione correctamente.
 |
(EXTRA): ¿Cuál es la longitud media de las palabras del texto que has escrito? “¿Hola cómo va?” debería devolver (4+3+2) / 3 = 3

```python
def contador_palabras(text):
    
    ''' Función que dado un texto muestra las veces (frecuencia) en que aparece cada palabra
    y el promedio de longitud de las palabras encontradas'''
    
    from collections import Counter
    import string 
    
    # Validacion de tipo de datos de ingreso (str)
    if not isinstance(text, str):
        raise TypeError("Debes introducir un texto (tipo string)")
    
    # Validacion de ingreso de datos no vacio
    if text.strip() == '':
        raise ValueError("El texto está vacío. Debe contener al menos una palabra")

    # Pasar todas las palabras a minusculas, limpiar signos de puntuacion y generar lista de palabras
    puntuacion_español = '¡,¿'
    puntuacion_total = puntuacion_español + string.punctuation
    
    text = text.lower() 
    text = ''.join(c for c in text if c not in puntuacion_total)
    
    if text == '':
        raise ValueError ('El texto no puede contener solo signos de puntuación')

    palabras = text.split()
    
    # Frecuencias y promedio
    
    # Genera un diccionario de frecuencias por palabra de la lista
    #frecuencias = Counter(palabras) 
    
    frecuencias = {}
    for palabra in palabras:
        if palabra not in frecuencias:
            frecuencias[palabra] = 1
        else:
            frecuencias[palabra] += 1    
    
    #Calcula el promedio de longitud de las palabras del texto
    #promedio = round(sum(len(palabra) for palabra in palabras) / len(palabras), 2)
    
    suma_letras= 0
    cantidad_palabras = 0

    for palabra,frecuencia in frecuencias.items():
        suma_letras += len(palabra)
        cantidad_palabras += 1
 
    promedio = suma_letras/cantidad_palabras

    return (
f"Frecuencias de palabras encontradas:\n {(frecuencias)}\n\n"
f"Promedio de longitud de palabras: {promedio}")
```

 Los valores y claves en el diccionario original son únicos; si éste no es el caso, la función

 Fíjate en los siguientes ejemplos para ver el output esperado. (ver pdf)

```python
def dic_inverted(dic):
    
    ''' Función que toma un diccionario y devuelve otro diccionario pero inverso:
        intercambiando sus items clave-valor'''
    
    from collections import defaultdict, Counter
        
    #Chequear la frecuencia de los values del diccionario ingresado
    # Si hay values no unicos, advertir por mensaje y presentar su lista
    valores = list(dic.values())
    
    #Comparar la longitud de la lista de valores vs la longitud con sus valores no repetidos (set)
    if len(valores) != len(set(valores)):
        print('Atención!: Hay valores repetidos en el diccionario ingresado.')
        print('El diccionario resultante tendra una lista de valores por cada clave.\n')
    
    # Generar un nuevo diccionario de frecuencias de la lista de valores
    contador_valores = Counter(valores)      
    
    # Generar una lista con los valores repetidos del diccionario contador_valores
    valores_repetidos = [k for k,v in contador_valores.items() if v > 1]

    print('Valores repetidos: ' , valores_repetidos)
              
    # Crear el diccionario vacio que permita tener valores de tipo lista
    invertido = defaultdict(list) 
    
    # Toma como key el value del diccionario original y como value/s los correspondientes key/s
    for k, v in dic.items():
        invertido[v].append(k)   
    
    return (print(f'Diccionario invertido:\n{dict(invertido)}'))
```

# 1. Contador y ordenador de palabras de un texto.

 El cliente quedó contento con el contador de palabras, pero ahora quiere leer archivos TXT y que calcule la frecuencia de cada palabra ordenadas dentro de las entradas habituales del diccionario según la letra con la que comienzan, es decir, las claves han de ir de la A a la Z y dentro de la A debemos ir de la A a la Z. Por ejemplo, para el archivo "tu_me_quieres_blanca.txt" la salida
 esperada sería:(ver archivo)

```python
def ruta_a_texto(ruta_archivo):
    
    ''' Esta Funcion abre y lee un archivo desde su ubicacion y devuelve un texto (str) '''
    
    try:
        with open(ruta_archivo, 'r', encoding='utf-8') as f:
            return f.read()
        
    except FileNotFoundError:
        print('Archivo no encontrado')
        
    except Exception as e:
        print(f'Ocurrió un error inesperado: {e}')
```

```python
def contador_ordenador_texto(text): 
    
    ''' Esta Funcion recibe un texto (string) y devuelve un diccionario alfabeticamente ordenado
        de las frecuencias de las palabras encontradas'''
    
    import string
    from collections import defaultdict
    
    # Validacion de tipo de datos de ingreso (str)
    if not isinstance(text, str):
        raise TypeError("Debes introducir un archivo de texto en formato 'string'")
    
    # Validacion de ingreso de datos no vacio
    if text.strip() == '':
        raise ValueError("Texto vacío: se espera una cadena con al menos una palabra o carácter no vacío")
            
    # Unificar todas las palabras en minusculas, limpiar puntuacion, generar lista de palabras limpias
    
    puntuacion_español = '¡¿'
    puntuacion_total = string.punctuation + puntuacion_español

    text = text.lower() 
    text = ''.join(c for c in text if c not in puntuacion_total) 
    palabras = text.split() 
    
    # Crear diccionarios anidados. El de dentro cuenta frecuencias de numeros enteros (int)
    # Con esta función anónima lambda, se anida automáticamente un diccionario dentro del otro

    resultado = defaultdict(lambda: defaultdict(int)) 
    
    for palabra in palabras:
        letra_inicial = palabra[0] 
        resultado[letra_inicial][palabra] += 1
    
    # Ordenar el diccionario por clave "externa" (letra inicial) 
    # Y por palabra:frecuencia del diccionario interno
    
    resultado_ordenado = {letra: dict(sorted(palabras.items()))
    for letra, palabras in sorted(resultado.items())} 
        
    return resultado_ordenado
```

# 2. Conversión de tipos de datos.

 El cliente recibe una lista de datos y necesita generar dos listas, la primera en la que estarán todos los elementos que se pudieron convertir en flotantes y la otra donde están los elementos que no pudieron convertirse.
 [ '1.3', 'one' , '1e10' , 'seven', '3-1/2', ('2',1,1.4,'not-a-number'), [1,2,'3','3.4']]

```python
def clasificacion_datos(lista):
    
    ''' Esta Funcion recibe una lista de datos y las divide en 2: una lista 
    con aquellos datos que se han podido transformar a numero flotantes y aquellos que no'''
    
    # Inicializar las 2 listas vacias
    flotantes = []
    no_flotantes = []

    for elem in lista:
            try:
            # Eliminar espacios en blanco al principio y final de cada elemento de la lista ingresada
                elem_limpio = str(elem).strip()

            # Prueba convertir a float
                flotantes.append(float(elem_limpio))
            
        # Si no lo puede convertir, levanta el error/excepcion y lo envia a la
        #lista no_flotantes
        
            except ValueError:
                print(f"No se pudo convertir '{elem}' : valor no válido")
                no_flotantes.append(elem)
            except Exception as e:
                print(f"Error inesperado con '{elem}': {e}")
                no_flotantes.append(elem)

    resultado = (f"Elementos a Flotantes:\n{flotantes}\n"
                f"Elementos No Flotantes:\n{no_flotantes}")
    
    return resultado
```

```python
lista_a_clasificar = (['1.3', 'one' , '1e10' , 'seven', '3-1/2', ('2',1,1.4,'not-a-number'), [1,2,'3','3.4']])
float_nofloat = clasificacion_datos(lista_a_clasificar)
print(float_nofloat)
```

# 1. Generador de contraseñas

Explora el funcionamiento del módulo random de la librería numpy. 
 En este punto, el cliente ha detectado un problema con las contraseñas que utilizan sus trabajadores. Asdf1234, fechas de cumpleaños o similares. Para solucionarlo, nos ha encargado una función de Python que genere contraseñas más seguras. La función debe depender de los siguientes parámetros:  

 longitud (int): Longitud de la contraseña
 mayúsculas (bool = True): Si tiene que aparecer mayúsculas
 minúsculas (bool = True): Si tiene que aparecer minúsculas
 números (bool = True): Si tiene que aparecer números
 signos (bool = False): Si tiene que aparecer caracteres especiales (,-$? o similares)

Así pues, si ejecutamos la función de la siguiente forma:

 crear_contraseña(10, True, True, True, True)
 Deberíamos obtener un output (que debemos poder guardar) del estilo:
 9Er,5Vn8P$

 Asegúrate de que se cumplan todos los criterios, y que estas contraseñas sean realmente aleatorias.
 (EXTRA) Explora cómo podríamos hacer que la función copiara la contraseña automáticamente en el portapapeles del ordenador (como si la hubiéramos seleccionado y hecho ctrl+copy).

```python
def crear_contraseñas1(largo=10, mayusculas=True, minusculas=True, numeros=True, signos=True):
    
    ''' Esta Funcion genera una contraseña 'aleatoria' 
    teniendo en cuenta la selección de los parametros ingresados'''
        
    # Del modulo string seleccionamos y guardamos los tipos de elementos que
    # podran ser parte de la contraseña en relacion con los parametros seleccionados
    
    import random
    import string  
     
    mayus = string.ascii_uppercase
    minus = string.ascii_lowercase
    nums = string.digits
    simbol = string.punctuation

    caracteres_bolsa = ''
    obligatorios = []

    if mayusculas:
        caracteres_bolsa += mayus
        obligatorios.append(random.choice(mayus))
    if minusculas:
        caracteres_bolsa += minus
        obligatorios.append(random.choice(minus))
    if numeros:
        caracteres_bolsa += nums
        obligatorios.append(random.choice(nums))
    if signos:
        caracteres_bolsa += simbol
        obligatorios.append(random.choice(simbol))

    if caracteres_bolsa == '':
        raise ValueError("Debe seleccionar al menos un tipo de carácter para generar la contraseña.")

    #if largo < len(obligatorios):
        #raise ValueError("El largo es demasiado corto para incluir todos los tipos activados.")

    # Si el largo de la contraseña es menor a la cantidad de caracteres obligatorios, no alcanza el espacio mínimo.
    if largo < len(obligatorios):
        raise ValueError("El largo de la contraseña es demasiado corto para incluir todos los tipos activados.")

    # Si el largo del a contraseña es mayor a la cantidad de caracteres obligatorios,
    # calcular la diferencia adicional y elegir aleatoriamente de caracteres_bolsa
    resto = largo - len(obligatorios)
    adicionales = [random.choice(caracteres_bolsa) for n in range(resto)]

    # Unir los obligatorios + adicionales, y desordenar para mayor aleatoriedad
    total = obligatorios + adicionales
    random.shuffle(total)
    contraseña = ''.join(total)
    # Devolver la contraseña final como string
    return contraseña
```

# 2. Procesamiento de datos simple

 Una compañera de trabajo nos ha pedido un favor, aprovechando que sabe que estamos aprendiendo a programar. Tiene un histórico de partidos de fútbol catalán en un fichero, donde se almacenan los nombres de los equipos y los resultados. Necesita que procesemos los datos de forma automática, para extraer los resultados que necesita.
 
 Utiliza el archivo "historic_partits.txt"
 Necesita un programa que devuelva:
 
 El número total de goles que ha marcado cada equipo.
 El nombre del equipo más goleador.
 El nombre del equipo más goleado
 La clasificación global (cada victoria: 3 pts, empate 1 pts, derrota 0 pts)
 Como referencia, los equipos con más puntos serían los siguientes

```python
def analisis_partidos(ruta_archivo):
    
    with open(ruta_archivo, 'r', encoding='utf-8') as f:
        partidos = [linea.strip().split('\t') for linea in f] #limpia y separa cada linea
    
    goles_favor = {}          
    goles_contra = {}        
    puntos = {}          
    
    for partido in partidos:
        equipo1, marcador, equipo2 = partido
        gol1, gol2 = map(int, marcador.split('-')) #separa los goles de cada equipo por el guion
                                                    # y los mapea equipo1-gol1, equipo2-gol2

        # En cada diccionario inicializa con key=equipo y valor cero si no están aún en los diccionarios
        for equipo in (equipo1, equipo2):
            goles_favor.setdefault(equipo, 0)
            goles_contra.setdefault(equipo, 0)
            puntos.setdefault(equipo, 0)

        # Sumar goles a favor y en contra
        goles_favor[equipo1] += gol1
        goles_contra[equipo1] += gol2
        goles_favor[equipo2] += gol2
        goles_contra[equipo2] += gol1

        # Asignar puntos segun la regla
        if gol1 > gol2:
            puntos[equipo1] += 3
        elif gol1 < gol2:
            puntos[equipo2] += 3
        else:
            puntos[equipo1] += 1
            puntos[equipo2] += 1

    return (
    f'Goles a Favor: {goles_favor}\n\n'
    f'Goles en Contra: {goles_contra}\n\n'
    f'Puntos: {puntos}')
   
```

```python
ruta = 'historic_partits.txt'
resultados = analisis_partidos(ruta)
print(resultados)
```
